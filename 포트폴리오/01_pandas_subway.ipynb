### 속성 뽑아내기

```python
subway.사용일자
```

```python
subway['사용일자']
```

```python
subway[ ['승차총승객수','하차총승객수'] ]
```

```python
data = subway[ ['승차총승객수','하차총승객수'] ].values
```

```python
from sklearn.linear_model import LinearRegression

X = data[:,0].reshape(-1,1)
y = data[:,1]

model = LinearRegression()
model.fit(X, y)
model.coef_, model.intercept_
```

```python
model.score(X, y)
```

```python
plt.scatter(data[:,0], data[:,1], alpha=0.1)
plt.plot([0,120000], [model.intercept_, 120000*model.coef_+model.intercept_], 'r-')
```
