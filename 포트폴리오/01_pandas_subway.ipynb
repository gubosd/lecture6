### Numpy 어레이로 변환 (values)

```python
data = subway[ ['승차총승객수','하차총승객수'] ].values
```

```python
from sklearn.linear_model import LinearRegression

X = data[:,0].reshape(-1,1)
y = data[:,1]

model = LinearRegression()
model.fit(X, y)
model.coef_, model.intercept_
```

```python
model.score(X, y)
```

```python
plt.scatter(data[:,0], data[:,1], alpha=0.1)
plt.plot([0,120000], [model.intercept_, 120000*model.coef_+model.intercept_], 'r-')
```

### 피벗 테이블
- pivot_table()

```python
pd.pivot_table(subway, values='승차총승객수', index='사용일자', columns='노선명', aggfunc=np.mean)
```

```python
p = pd.pivot_table(subway, values='승차총승객수', index='사용일자', columns='노선명', aggfunc=np.mean)

wday = {0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}
p.insert(0, '요일', pd.to_datetime(p.index, format='%Y%m%d').dayofweek.map(wday))
p
```
