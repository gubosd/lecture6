```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
cancer.feature_names
```

```python
#X = cancer.data[:,0].reshape(-1,1)
X = cancer.data[:, [0]]
y = cancer.target
```

```python
plt.scatter(X, y, c=y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
model.score(X, y)
```

```python
w = model.coef_[0]
b = model.intercept_

w, b
```

```python
plt.scatter(X, y, c=y)
plt.plot([5,30], [w*5+b, w*30+b])
```

```python
X = cancer.data[:,[0]]
y = cancer.data[:, 1]
```

```python
plt.scatter(X, y)
```

```python
model = LinearRegression()
model.fit(X, y)
print(model.score(X, y))

w = model.coef_[0]
b = model.intercept_

plt.scatter(X, y)
plt.plot([5,30], [w*5+b, w*30+b], 'r--')
```
