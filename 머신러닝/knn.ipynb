```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.data.shape
```

```python
cancer.target.shape
```

```python
X = cancer.data
y = cancer.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model.fit(X, y)
model.score(X, y)
```

```python
예측값 = model.predict(X)
(예측값!=y).sum()
```

```python
오류점들 = X[예측값!=y]
```

```python
plt.scatter(오류점들[:,0], 오류점들[:,27], s=100, c='red')
plt.scatter(X[:,0], X[:,27], c=y, alpha=0.2)
```

```python
from sklearn.model_selection import train_test_split
X_훈련, X_검증, y_훈련, y_검증 = train_test_split(X, y)
```

```python
X_훈련.shape, X_검증.shape
```

```python
model2 = KNeighborsClassifier(3)
model2.fit(X_훈련, y_훈련)

model2.score(X_훈련, y_훈련), model2.score(X_검증, y_검증)
```

```python
X = cancer.data[:, [1,10]]
y = cancer.target
print(X.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, alpha=0.2)
plt.axis('scaled')
```

```python
model3 = KNeighborsClassifier(10)
model3.fit(X, y)
model3.score(X, y)
```

```python
X[:5]
```

```python
import mglearn

#plt.figure(figsize=[12,10])

mglearn.plots.plot_2d_classification(model3, X, fill=True, eps=0.5, alpha=0.4)
mglearn.discrete_scatter(X[:,0], X[:,1], y)

plt.axis('scaled')
```

```python
X_정규화 = (X - X.mean(axis=0))/X.std(axis=0)
X_정규화[:5]
```

```python
plt.scatter(X_정규화[:,0], X_정규화[:,1], c=y, alpha=0.2)
plt.axis('scaled')
```

```python
model4 = KNeighborsClassifier()
model4.fit(X_정규화, y)
model4.score(X_정규화, y)
```

```python
import mglearn

mglearn.plots.plot_2d_classification(model4, X_정규화, fill=True, eps=0.5, alpha=0.4)
mglearn.discrete_scatter(X_정규화[:,0], X_정규화[:,1], y)

plt.axis('scaled')
```
