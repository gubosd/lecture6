```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
from sklearn.svm import LinearSVC
```

```python
X = cancer.data
y = cancer.target
```

```python
svc = LinearSVC()
svc.fit(X, y)
svc.score(X, y)
```

```python
X_norm = (X - X.mean(axis=0))/X.std(axis=0)
```

```python
svc = LinearSVC()
svc.fit(X_norm, y)
svc.score(X_norm, y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X, y)
lr.score(X, y)
```

```python
lr = LogisticRegression()
lr.fit(X_norm, y)
lr.score(X_norm, y)
```

```python
y_svc = svc.predict(X_norm)
y_lr = lr.predict(X_norm)
```

```python
np.where(y_svc!=y), np.where(y_lr!=y)
```

```python
idx = np.where(y_svc!=y)[0]

plt.scatter(X[idx][:,0], X[idx][:,27], c='red', s=100)
plt.scatter(X[:,0], X[:,27], c=y, alpha=0.2)
```

```python
idx = np.where(y_lr!=y)[0]

plt.scatter(X[idx][:,0], X[idx][:,27], c='red', s=100)
plt.scatter(X[:,0], X[:,27], c=y, alpha=0.2)
```

```python
lr.predict_proba(X_norm)[idx]
```

```python
lr.decision_function(X_norm)[idx]
```
