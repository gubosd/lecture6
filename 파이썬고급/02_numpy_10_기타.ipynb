### 항목별로 적용되는 함수 만들기

```python
def myf(n):
    if n>0:
        return 1
    else:
        return -1
    
myf(2)
```

```python
a = np.random.randn(3,3)
print(a)

np.vectorize(myf)(a)
```

### 행렬 연산, 벡터 연산

-  벡터곱(norm) : 같은 축의 값을 곱해서 모두 더한다

```python
a = np.array([1,2])
b = np.array([2,1])

(a*b).sum(), (a*a).sum()
```

- 행렬곱 : AB 에서 A의 열의 갯수와 B의 행의 갯수가 같을 때
- 예를 들어 100X4 와 4X2 행렬을 곱하면 100X2 행렬이 된다

```python
a = np.arange(6).reshape(2,3)
b = np.arange(6).reshape(3,2)
display(a,b)
```

```python
a@b
```

```python
np.dot(a,b)
```

```python
np.matmul(a,b)
```

### np.meshgrid()
- x축과 y축의 구간값을 알려주면, x축의 값과 y축의 값을 가지는 행렬 2개를 만들어 준다

```python
np.meshgrid([1,2,3], [10,20,30])
```

```python
X, Y = np.meshgrid(np.arange(-10,10,0.1), np.arange(-10,10,0.1))
#Z = X*Y
#Z = X**2 + Y**2
Z = (X*Y)*np.sin(X*Y)

plt.imshow(Z, cmap='gray')
plt.colorbar()
```
