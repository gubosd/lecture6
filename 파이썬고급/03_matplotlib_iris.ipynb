# Iris 데이터 시각화 및 분석

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 아이리스 데이터 불러오기

```python
s = open('iris.csv').readline()
header = s.strip().split(',')[:-1]
header
```

```python
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
iris = np.loadtxt('iris.csv', delimiter=',', skiprows=1,
                  converters={4: lambda s: labels.index(s.decode())})
```

```python
display(iris.shape, iris[:5])
```

```python
X = iris[:,:4]
y = iris[:,4]
```

```python
display(X.shape, y.shape)
display(X[:5], y)
```

### 박스플롯
- 전체 속성의 전반적인 범위를 알 수 있다.

```python
plt.boxplot(X)
#plt.xticks(range(1,5), header, rotation=90)
plt.xticks(range(1,5), ['SL','SW','PL','PW'], rotation=90)
```

### 꺽은선그래프 (plot)
- 개개의 샘플들이 어떤 값을 가지는지 속성별로 그려보자

```python
plt.plot(iris[:,0])
plt.title('Sepal Length')
```

```python
plt.plot(iris[:,1])
plt.title('Sepal Width')
```

```python
plt.plot(iris[:,2])
plt.title('Petal Length')
```

```python
plt.plot(iris[:,3])
plt.title('Petal Width')
plt.vlines([50,100], 0, 3, linestyles=':')
plt.hlines([0.8, 1.6], 0, 150, linestyles=':', colors='red')
plt.text(20,2.5,'Setosa')
```

```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(iris[:,i])
    plt.title(features[i])
```

### 히스토그램
- 각 속성별로 어디에 샘플들이 많이 분포하고 있는지를 알 수 있다

```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.hist(iris[:,i], bins=20)
    plt.title(features[i])
```

- Petal Length 에서 2.2 를 경계로 왼쪽 섬에 있는 친구들이 누구인지 알고 싶어진다

```python
iris[iris[:,2]<2.2]
```

```python
plt.hist(iris[:,2][:50], bins=np.arange(1,7,0.2), alpha=0.2)
plt.hist(iris[:,2][50:100], bins=np.arange(1,7,0.2), alpha=0.2)
plt.hist(iris[:,2][100:], bins=np.arange(1,7,0.2), alpha=0.2)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
```

```python
for i in range(4):
    plt.subplot(2,2,i+1)
    r1 = iris[:,i].min()-0.1
    r2 = iris[:,i].max()+0.1
    
    plt.hist(iris[:,i][:50], bins=np.arange(r1,r2,0.2), alpha=0.2)
    plt.hist(iris[:,i][50:100], bins=np.arange(r1,r2,0.2), alpha=0.2)
    plt.hist(iris[:,i][100:], bins=np.arange(r1,r2,0.2), alpha=0.2)
    #plt.legend(['Setosa', 'Versicolor', 'Virginica'])
    plt.title(features[i])
```

### 산점도
- 두 속성 간의 상관관계를 알아본다.

```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
col1 = 2
col2 = 3

plt.scatter(X[:,col1], X[:,col2], c=y, alpha=0.5)
plt.xlabel(features[col1])
plt.ylabel(features[col2])
plt.title('Petal Length vs Petal Width')
```

```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

n = 1

plt.figure(figsize=[12,12])

for col1 in range(3):
    for col2 in range(col1+1,4):
        plt.subplot(3,2,n)
        plt.scatter(X[:,col1], X[:,col2], c=y, alpha=0.5)
        plt.xlabel(features[col1])
        plt.ylabel(features[col2])
        n = n+1
```
